```python
# NYC Citibike Program 2019 - get May 2019 - July 2019 tripdata
import pandas as pd
import datetime
import numpy as np
import os
```

```python
# Get NYC Citibike tripdata csv files from 201905 to 201907
filepath_may = os.path.join("Resources" ,"JC-201905-citibike-tripdata.csv")
filepath_jun = os.path.join("Resources" ,"JC-201906-citibike-tripdata.csv")
filepath_jul = os.path.join("Resources" ,"JC-201907-citibike-tripdata.csv")

nycbike_2019may = pd.read_csv(filepath_may)
nycbike_2019jun = pd.read_csv(filepath_jun)
nycbike_2019jul = pd.read_csv(filepath_jul)
```

```python
nycbike_2019may.head(10)
```

```python
nycbike_may_jun_jul = pd.concat([nycbike_2019may,nycbike_2019jun,nycbike_2019jul])
```

```python
# Output concatenated 2019 May-July citibike tripdata to CSV file 
filepath_2019_may_jul = os.path.join("Resources" ,"JC-2019-may-jul-citibike-tripdata1.csv")
nycbike_may_jun_jul.to_csv(filepath_2019_may_jul, index=False)
nycbike_may_jun_jul.tail(10)
```
